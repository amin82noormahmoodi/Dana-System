```python
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import FAISS
from langchain.chat_models import ChatOpenAI
```

# Behran

```python
with open(r"data\behran.txt", "r", encoding="utf-8") as file:
    behran = file.read()
```

```python
behran_splits = behran.split("\n\n")
```

```python
OPENAI_API_KEY = "OPENAI_API_KEY"










```

```python
embedding_function = OpenAIEmbeddings(openai_api_key = OPENAI_API_KEY)
```

```python
behran_db = FAISS.from_texts(
    behran_splits,
    embedding_function)
```

```python
behran_db.save_local(r"vectordb\behran")
```

```python
def build_save_vectordb(text_file_address, save_file_address):
    with open(text_file_address, "r", encoding="utf-8") as file:
        text = file.read()
    splits = text.split("\n\n")
    embedding_function = OpenAIEmbeddings(
        openai_api_key = OPENAI_API_KEY
    )
    db = FAISS.from_texts(
        splits,
        embedding_function
    )
    db.save_local(save_file_address)
```

# Sina

```python
build_save_vectordb(r"data\sina.txt", r"vectordb\sina")
```

# Irantire

```python
build_save_vectordb(r"data\irantire.txt", r"vectordb\irantire")
```

# Companies

```python
build_save_vectordb(r"data\companies.txt", r"vectordb\companies")
```

```python
embedding_function = OpenAIEmbeddings(
    openai_api_key = OPENAI_API_KEY
)
db = FAISS.load_local(r"vectordb\behran",
                      embeddings=embedding_function,
                      allow_dangerous_deserialization=True)
retriever = db.as_retriever()
llm = ChatOpenAI(
    model = "gpt-4o-mini",
    openai_api_key = OPENAI_API_KEY,
    temperature = 0.0)
```

```python
docs = retriever.invoke("معرفی بهرن", k=5)
```

```python
docs_content = []
for elemenet in docs:
    docs_content.append(elemenet.page_content)
```

```python
docs_content
```

```python
query = "معرفی بهرن"
prompt = f"""

این کوئری کاربر هستش
{query}

این اطلاعاتی هستش که بازیابی شده
{docs_content}

بر اساس اطلاعات بازیابی شده، پاسخ به کوئری کاربر بنویس
"""

response = llm.invoke(prompt).content
```
